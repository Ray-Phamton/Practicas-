```python

#importando librería datetime para el ordenado por fecha
from datetime import datetime

# Lista de 24 productos de ejemplo
productos = [

  {"marca": "Toyota", "modelo": "Corolla", "ano": 2020, "nombre": "Batería", "precio": 80, "fecha": "08/03/25", "calificacion": 4.5, "ventas": 120},

  {"marca": "Toyota", "modelo": "Camry", "ano": 2021, "nombre": "Batería", "precio": 90, "fecha": "16/04/25", "calificacion": 4.2, "ventas": 95},

  {"marca": "Chevrolet", "modelo": "Silverado", "ano": 2020, "nombre": "Batería", "precio": 75, "fecha": "23/04/25", "calificacion": 3.8, "ventas": 150},

  {"marca": "Ford", "modelo": "Fiesta", "ano": 2020, "nombre": "Batería", "precio": 130, "fecha": "30/04/25", "calificacion": 4.7, "ventas": 80},

  {"marca": "Toyota", "modelo": "Corolla", "ano": 2023, "nombre": "Batería", "precio": 300, "fecha": "07/05/25", "calificacion": 4.9, "ventas": 55},

  {"marca": "Chevrolet", "modelo": "Silverado", "ano": 2019, "nombre": "Batería", "precio": 125, "fecha": "14/05/25", "calificacion": 4.9, "ventas": 110},

  {"marca": "Ford", "modelo": "Fiesta", "ano": 2018, "nombre": "Batería", "precio": 65, "fecha": "21/05/25", "calificacion": 4.0, "ventas": 180},

  {"marca": "Toyota", "modelo": "Tacoma", "ano": 2022, "nombre": "Batería", "precio": 50, "fecha": "28/05/25", "calificacion": 4.3, "ventas": 220},

  {"marca": "Chevrolet", "modelo": "Silverado", "ano": 2021, "nombre": "Batería", "precio": 85, "fecha": "04/06/25", "calificacion": 3.9, "ventas": 70},

  {"marca": "Ford", "modelo": "Explorer", "ano": 2019, "nombre": "Batería", "precio": 70, "fecha": "11/06/25", "calificacion": 4.6, "ventas": 250},

  {"marca": "Toyota", "modelo": "4Runner", "ano": 2020, "nombre": "Batería", "precio": 40, "fecha": "18/06/25", "calificacion": 4.1, "ventas": 190},

  {"marca": "Chevrolet", "modelo": "Traverse", "ano": 2023, "nombre": "Batería", "precio": 55, "fecha": "25/06/25", "calificacion": 3.7, "ventas": 135},

  {"marca": "Ford", "modelo": "Mustang", "ano": 2017, "nombre": "Batería", "precio": 190, "fecha": "02/07/25", "calificacion": 4.8, "ventas": 45},

  {"marca": "Toyota", "modelo": "Prius", "ano": 2019, "nombre": "Batería", "precio": 115, "fecha": "09/07/25", "calificacion": 4.4, "ventas": 100},

  {"marca": "Chevrolet", "modelo": "Blazer", "ano": 2022, "nombre": "Batería", "precio": 140, "fecha": "16/07/25", "calificacion": 3.6, "ventas": 60},

  {"marca": "Toyota", "modelo": "Corolla", "ano": 2020, "nombre": "Batería", "precio": 30, "fecha": "26/12/25", "calificacion": 2.1, "ventas": 20},

  {"marca": "Toyota", "modelo": "Hilux", "ano": 2022, "nombre": "Batería", "precio": 110, "fecha": "15/03/25", "calificacion": 4.8, "ventas": 95},

  {"marca": "Toyota", "modelo": "Corolla", "ano": 2021, "nombre": "Batería", "precio": 95, "fecha": "20/03/25", "calificacion": 4.6, "ventas": 110},

  {"marca": "Toyota", "modelo": "Yaris", "ano": 2023, "nombre": "Batería", "precio": 90, "fecha": "25/03/25", "calificacion": 4.7, "ventas": 105},

  {"marca": "Chevrolet", "modelo": "Silverado", "ano": 2020, "nombre": "Batería", "precio": 85, "fecha": "01/04/25", "calificacion": 4.5, "ventas": 125},

  {"marca": "Chevrolet", "modelo": "Cruze", "ano": 2019, "nombre": "Batería", "precio": 75, "fecha": "05/04/25", "calificacion": 4.3, "ventas": 140},

  {"marca": "Chevrolet", "modelo": "Silverado", "ano": 2022, "nombre": "Batería", "precio": 120, "fecha": "10/04/25", "calificacion": 4.9, "ventas": 80},

  {"marca": "Chevrolet", "modelo": "Onix", "ano": 2024, "nombre": "Batería", "precio": 90, "fecha": "12/04/25", "calificacion": 4.7, "ventas": 115},

  {"marca": "Ford", "modelo": "Fiesta", "ano": 2020, "nombre": "Batería", "precio": 70, "fecha": "18/03/25", "calificacion": 4.2, "ventas": 150},

]

#constantes

MENU_FILTRADO = [
    "FILTRAR POR:".center(50),
    "1. Marca".center(45),
    "2. Modelo".center(47),
    "3. Año".center(44),
    "4. Marca y Modelo".center(56),
    "5. Marca y Año".center(52),
    "6. Modelo y Año".center(54),
    "7. Marca, Modelo y Año".center(59),
    "8. Salir".center(46)
]

MENU_ORDENADO = [
    "ORDENAR POR:".center(50),
    "1. precio".center(48),
    "2. fecha".center(45),
    "3. calificación".center(53),
    "4. ventas".center(48)
]

MENU_MARCAS =[
    "indique la marca según la lista:".center(69),
    "Toyota".center(44),
    "Chevrolet".center(47),
    "Ford".center(42)
]

SEPARADORES = "-" * 40

# Función de filtrado

def filtrar_productos(marca=None,  modelo=None, ano=None):

    resultados = productos

    if marca:
        resultados = [p for p in resultados if p["marca"] == marca]
    if modelo:
        resultados = [p for p in resultados if p["modelo"] == modelo]
    if ano:
        resultados = [p for p in resultados if p["ano"] == ano]
    return resultados

#ordenamiento de menor a mayor

def merge_sort(lista, clave):
    if len(lista) <= 1:
        return lista

    medio = len(lista) // 2
    izquierda = merge_sort(lista[:medio], clave)
    derecha = merge_sort(lista[medio:], clave)

    return merge(izquierda, derecha, clave)

def merge(izq, der, clave):
    resultado = []
    i = j = 0

    while i < len(izq) and j < len(der):
        if izq[i][clave] <= der[j][clave]:
            resultado.append(izq[i])
            i += 1
        else:
            resultado.append(der[j])
            j += 1

    # Agregar lo que queda
    resultado.extend(izq[i:])
    resultado.extend(der[j:])
    return resultado

#ordenamiento de mayor a menor

def merge_sort_descendiente(lista, clave):
    if len(lista) <= 1:
        return lista

    medio = len(lista) // 2
    izquierda = merge_sort_descendiente(lista[:medio], clave)
    derecha = merge_sort_descendiente(lista[medio:], clave)

    return merge_descendiente(izquierda, derecha, clave)

def merge_descendiente(izq, der, clave):
    resultado = []
    i = j = 0

    while i < len(izq) and j < len(der):
        if izq[i][clave] >= der[j][clave]:
            resultado.append(izq[i])
            i += 1
        else:
            resultado.append(der[j])
            j += 1

    resultado.extend(izq[i:])
    resultado.extend(der[j:])
    return resultado

#ordenamiento por fecha

def merge_sort_fecha(lista, clave):
    if len(lista) <= 1:
        return lista

    medio = len(lista) // 2
    izquierda = merge_sort_fecha(lista[:medio], clave)
    derecha = merge_sort_fecha(lista[medio:], clave)

    return merge_fecha(izquierda, derecha, clave)

def merge_fecha(izq, der, clave):
    resultado = []
    i = j = 0

    while i < len(izq) and j < len(der):
        fecha_i = datetime.strptime(izq[i][clave], "%d/%m/%y")
        fecha_j = datetime.strptime(der[j][clave], "%d/%m/%y")

        if fecha_i >= fecha_j:
            resultado.append(izq[i])
            i += 1
        else:
            resultado.append(der[j])
            j += 1

    resultado.extend(izq[i:])
    resultado.extend(der[j:])
    return resultado

#Sistema de filtrado.

#Con un filtro:
def mostrar_datos(opcion = None):
    if opcion == 1:
        print("\n".join(MENU_MARCAS))
        Marca = input("Marca:")
        return filtrar_productos(marca= Marca)

    elif opcion == 2:
        Modelo= input("indique el modelo que busca(Corolla,Silverado,Fiesta):")

        return filtrar_productos(modelo=Modelo)


    elif opcion == 3:
        Ano = int(input("indique el año del producto(2020):"))
        return filtrar_productos(ano=Ano)

#Con 2 filtros:

    elif opcion == 4:
        print("\n".join(MENU_MARCAS))
        Marca = input("Marca:")
        Modelo= input("indique el modelo que busca(Corolla,Silverado,Fiesta):")


        return filtrar_productos(marca =Marca, modelo=Modelo)

    elif opcion == 5:
        print("\n".join(MENU_MARCAS))
        Marca = input("Marca:")
        Ano =int(input("indique el año del producto(2020):"))
        return filtrar_productos(marca =Marca, ano=Ano)

    elif opcion == 6:
        Modelo= input("indique el modelo que busca(Corolla,Silverado,Fiesta):")
        Ano = int(input("indique el año del producto(2020):"))
        return filtrar_productos(modelo =Modelo, ano=Ano)

#Con 3 filtros:

    elif opcion == 7:
        print("\n".join(MENU_MARCAS))
        Marca = input("Marca:")
        Modelo= input("indique el modelo que busca(Corolla,Silverado,Fiesta):")
        Ano =int(input("indique el año del producto(2020):"))
        return filtrar_productos(marca=Marca, modelo=Modelo, ano=Ano)

#funcion para determinar el ordenamiento
#esta funcion existe para simular botones la idea principal es que el usuario no deba ingresar nada.

def ordenar_por(opcion):
    if opcion == 1:
        return "precio"
    elif opcion == 2:
        return "fecha"
    elif opcion == 3:
        return "calificacion"
    elif opcion == 4:
        return "ventas"
    else:
        print ("ingrese un valor válido")

#ejecución

while True:
    print("\n".join(MENU_FILTRADO))

    opcion_filtrado = int(input ("ingrese cómo desea filtrar(1-8): "))

    if opcion_filtrado == 8:
        print("Gracias por preferirnos! :D")
        break
    print()
    print("\n".join(MENU_ORDENADO))

    opcion_ordenado = ordenar_por(int(input("ingrese como desea ordenar(1-4): ")))

    if opcion_ordenado == "precio":
        filtrados = mostrar_datos(opcion_filtrado)
        ordenados = merge_sort(filtrados, opcion_ordenado)

    elif opcion_ordenado == "fecha":
        filtrados = mostrar_datos(opcion_filtrado)
        ordenados = merge_sort_fecha(filtrados, opcion_ordenado)

    else:
        filtrados = mostrar_datos(opcion_filtrado)
        ordenados = merge_sort_descendiente(filtrados, opcion_ordenado)

    print(" ");
    for diccionario in ordenados:
        for clave, valor in diccionario.items():
            print(f"{clave}: {valor}".center(50))
        print(SEPARADORES.center(50))
```
